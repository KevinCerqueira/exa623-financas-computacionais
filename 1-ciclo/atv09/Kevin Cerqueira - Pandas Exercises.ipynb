** Import pandas and read in the banklist.csv file into a dataframe called banks. **

** What are the column names? **

** How many States (ST) are represented in this data set? **

```python
num_states = banks['ST'].nunique()
print(num_states)
```

** Get a list or array of all the states in the data set. **

```python
states = banks['ST'].unique()
print(states)
```

** What are the top 5 states with the most failed banks? **

```python
top_5_states = banks['ST'].value_counts().head(5)
print(top_5_states)
```

** What are the top 5 acquiring institutions? **

```python
top_5_acquirers = banks['Acquiring Institution'].value_counts().head(5)
print(top_5_acquirers)
```

** How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?**

```python
state_bank_of_texas_acquisitions = banks[banks['Acquiring Institution'] == 'State Bank of Texas']

num_acquired_by_sbt = state_bank_of_texas_acquisitions.shape[0]

num_acquired_in_tx = state_bank_of_texas_acquisitions[state_bank_of_texas_acquisitions['ST'] == 'TX'].shape[0]

print(state_bank_of_texas_acquisitions)
```

** What is the most common city in California for a bank to fail in?**

```python
most_common_city_ca = banks[banks['ST'] == 'CA']['City'].value_counts().idxmax()
print(most_common_city_ca)
```

** How many failed banks don't have the word "Bank" in their name? **

```python
no_bank_in_name = banks[~banks['Bank Name'].str.contains("Bank")].shape[0]
print(no_bank_in_name)
```

** How many bank names start with the letter 's' ? **

```python
banks_starting_with_s = banks[banks['Bank Name'].str.startswith('S')].shape[0]
print(banks_starting_with_s)
```

** How many CERT values are above 20000 ? **

```python
cert_above_20000 = banks[banks['CERT'] > 20000].shape[0]
print(cert_above_20000)
```

** How many bank names consist of just two words? (e.g. "First Bank" , "Bank Georgia" )**

```python
two_word_banks = banks[banks['Bank Name'].str.split().apply(len) == 2].shape[0]
print(two_word_banks)
```
